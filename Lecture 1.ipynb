```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 02/18/18 Xinyi Y Zhou adapting MATLAB Quantization.m to Python code.

#  EE368/CS232 Digital Image Processing
# % Bernd Girod
# % Department of Electrical Engineering, Stanford University 

# % Script by Qiyuan Tian and David Chen
# % Quantization: how many bits per pixel?

# Load test image
img = cv2.imread('face.jpg',0);

# Loop over number of bits
for numOfBit in range(1, 9): 
    # Quantize to given number of bits
    numOfLevel = 2**numOfBit;
    levelGap = 256 / numOfLevel;
    quantizedImg = cv2.convertScaleAbs((np.ceil(img / levelGap) * levelGap - 1)); # quantization
   
    # Plot image
    plt.figure()
    plt.imshow(quantizedImg, cmap='gray');
    
    if numOfBit == 1:
        name = str(numOfBit) + '-bit';
    else:
        name = str(numOfBit) + '-bits';
    plt.title(name);
```
