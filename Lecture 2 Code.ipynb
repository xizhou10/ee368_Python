```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 02/18/18 Xinyi Y Zhou adapting MATLAB Quantization.m to Python code.

#  EE368/CS232 Digital Image Processing
# % Bernd Girod
# % Department of Electrical Engineering, Stanford University 

# % Script by Qiyuan Tian and David Chen
# % Quantization: how many bits per pixel?

# Load test image
img = cv2.imread('face.jpg',0);

plt.figure()

# Loop over number of bits
for numOfBit in range(1, 9): 
    # Quantize to given number of bits
    numOfLevel = 2**numOfBit;
    levelGap = 256 / numOfLevel;
    quantizedImg = cv2.convertScaleAbs((np.ceil(img / levelGap) * levelGap - 1)); # quantization
   
    # Plot image
    plt.subplot(2, 4, 9 - numOfBit)
    plt.imshow(quantizedImg, cmap='gray');
    
    if numOfBit == 1:
        name = str(numOfBit) + '-bit';
    else:
        name = str(numOfBit) + '-bits';
    plt.title(name);
```

```python
# 02/18/18 Xinyi Y Zhou adapting MATLAB Brightness_Contrast.m to Python code.

# % EE368/CS232 Digital Image Processing
# % Bernd Girod
# % Department of Electrical Engineering, Stanford University 

# % Script by Qiyuan Tian and David Chen
# % Brightness Adjustment and Contrast Adjustment

# Load test image
img = cv2.imread('parrot.jpg',0);

# Brightness adjustment by intensity scaling
scale = 1.2;
scaledImg = scale*img;

plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray');
plt.title('Original image');

plt.subplot(1, 2, 2)
plt.imshow(scaledImg, cmap='gray');
plt.title('Scaled image');

# Contrast adjustment by changing "gamma"
gamma = 1.5;
gammaImg = img**gamma;

plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray');
plt.title('Original image');

plt.subplot(1, 2, 2)
plt.imshow(gammaImg, cmap='gray');
plt.title('Gamma increased by 50%');
```
